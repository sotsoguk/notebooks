```python
import matplotlib.pyplot as plt
```

```python

maxNum = 1000
occurs = [0] * maxNum
prevSeen = [0] * maxNum
sequence = [0,0]
prevSeen[0] = 2
occurs[0] = 1
for i in range(3,maxNum+1):
    currElem = sequence[-1]
    occurs[currElem] =+ 1
    # New Element
    if (prevSeen[currElem] == 0):
        prevSeen[currElem] = i
        sequence.append(0)
    else:
        sequence.append(i - prevSeen[currElem])
        prevSeen[currElem] = i
        
    
```

```python
# look for missing number
missing = []
for i in range(0,maxNum):
    if (occurs[i] == 0):
        missing.append(i)
```

```python
plt.plot(sequence)
plt.title('vanEck-Sequence')
plt.show()
```
