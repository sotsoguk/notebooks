```python
maxNum = 10000
prevSeen = [0] * maxNum
sequence = [0,0]
prevSeen[0] = 2

for i in range(3,maxNum+1):
    currElem = sequence[-1]
    # New Element
    if (prevSeen[currElem] == 0):
        prevSeen[currElem] = i
        sequence.append(0)
    else:
        sequence.append(i - prevSeen[currElem])
        prevSeen[currElem] = i
        
    
```
